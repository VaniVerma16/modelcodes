```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating a sample dataset
np.random.seed(42) #to make sure that each time the same random data is generated
X = 2*np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100, 1)
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Sample Data")
plt.show()
```

```python
theta = np.zeros((2,1)) #initialising the theta values
```

```python
X_b = np.c_[np.ones((100, 1)), X] #letting theta-0 = 1 according to our assumption
```

```python
#mean squared error function for cost function
def mse(X,y,theta):
    m = len(y)
    h = X.dot(theta) #h(x) = X*theta, both are vectors
    cost = (1 / (2 * m)) * np.sum((h - y) ** 2) #sum up the (predictions-y)^2 values over the entire dataset
    return cost
```

```python
#grad desc function for minimising costs
def grad_desc(X,y,theta,alpha, iterations):
    m = len(y)
    costs=[]
    for i in range(iterations):
        h = X.dot(theta) #h = X*theta
        gradients = (1 / m) * X.T.dot(h - y) #calculation of grads
        theta -= alpha*gradients
        cost = mse(X, y, theta)
        costs.append(cost)
    return theta, costs
```

```python
#defining parameters
alpha = 0.1
iterations = 100

#what are values of theta after 100 iterations
theta, cost_history = grad_desc(X_b, y, theta, alpha, iterations)
print(f"Theta_0: {theta[0][0]}, Theta_1: {theta[1][0]}")
```

```python
#plot the regression line
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta), color="red")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.show()

#plot the cost history
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Over Time")
plt.show()
```
